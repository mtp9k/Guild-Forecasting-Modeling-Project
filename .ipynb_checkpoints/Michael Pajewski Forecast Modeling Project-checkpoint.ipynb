### For each program category count the number of distinct contacts (students) that are enrolled in courses.

Assumptions:

1. There are 5 program categories
    Lead Gen - Post Secondary,
    University,
    Credential,
    StraighterLine/Saylor,
    Lead Gen - Secondary, and 
    ELL Program

2. Students can only be enrolled in one category i.e. a student could not be enrolled in both a University and the ELL Program

3. Sf Contact ID is a unique student ID

```python
# SQL Query
myquery = """
SELECT Opportunity_Details.SfOpportunityProgramCategory as Program_Category, COUNT(Contact_Details.SfContactID) as Count_Contacts
FROM Opportunity_Details
    LEFT JOIN Contact_Details
    ON Opportunity_Details.SfOpportunityID = Contact_Details.SfOpportunityID
GROUP BY Opportunity_Details.SfOpportunityProgramCategory
ORDER BY Count_Contacts DESC;
"""

#Execute SQL Query
pd.read_sql_query(myquery, con=engine)
```

## Based on your best interpretation of the data, which application type has the most number of successful students? Keep this definition in mind for the Python portion below)

Assumptions:

1. From the distinct list below 

    Passing grades are A, A-, B, B+, B-, C, C-, C+,  Passed, S, SP 

    Failing grades are FA, W, F, D, D+, D-, Failed, WP, FW, IF, WA, I, U
2. Sf Contact ID is a unique ID representing an enrolment in a course by a student i.e. each student can have many Sf Course C ID's but each Sf Course C ID is unique

3. SfOpportunityApplicationTypeC from the Opportunity_Details is the application type


```python
#query to show distinct grades given
myquery2 = """
SELECT DISTINCT(SfCourseCFinalGradeC)
FROM Course_Details;
"""

#Execute SQL Query
results = pd.read_sql_query(myquery2, con=engine)
results = pd.DataFrame(results)
print('List of distinct grades')
print(results['SfCourseCFinalGradeC'].to_list())
```

```python
#Query to determin the number of students and percent pass rate for each application type
myquery3 = """
SELECT OD.SfOpportunityApplicationTypeC as Application_Type, 
        COUNT(CorD.SfCourseCFinalGradeC) as Number_Students,
-- Create AVG Case to define when students passed or failed a course (1 pass 0 failed) 
    AVG(CASE WHEN CorD.SfCourseCFinalGradeC = 'Passed' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'A' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'A-' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'B' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'B+' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'B-' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'C' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'C+' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'C-' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'S' THEN 1
        WHEN CorD.SfCourseCFinalGradeC = 'SP' THEN 1
        ELSE 0 
    END) AS Pass_Fail_Percentage
-- join the three tables together so we have  connection between Application Type and Final Grade C
    FROM Opportunity_Details as OD
        INNER JOIN
        Contact_Details AS ConD
        ON OD.SfOpportunityID = ConD.SfOpportunityID
        INNER JOIN
        Course_Details as CorD
        ON ConD.SfCourseCID = CorD.SfCourseCID
-- Filter for entries that have grades, removing blank
    WHERE CorD.SfCourseCFinalGradeC <> ''
    GROUP BY OD.SfOpportunityApplicationTypeC
-- Order By the percent of studetns who passed course
    ORDER BY Pass_Fail_Percentage DESC;
"""
```

From the table below we can see that all students in programs with the University of Denver passed the courses, but there is a extremely small sample size of 9 students. The low sample size makes it difficult to fully evaluate courses University of Denver.  96% of the courses taken at Western Governors University are passed but the institution still has a small total enrolment compared to the other institutions. Penn Foster High School and Guild Education have similar pass rates at 82% and 79% but Guild Education has significantly more students enrolled. My belief is that Penn Foster High School and Guild Education have the most successful students based on their pass fail percentage and their enrolment numbers.

## Data Cleaning

Assumptions carry over from the previous SQL work additional assumptions include: 
1. Blank SfCourseCFinalGradeC means there is no information on if the course was passed or failed. These will not be included in the model


```python
#using similar work from the previous problem to join together the 3 tables
myquery4 = """
SELECT *
-- join the three tables together so we have  connection between Application Type and Final Grade C
    FROM Opportunity_Details as OD
        INNER JOIN
        Contact_Details AS ConD
        ON OD.SfOpportunityID = ConD.SfOpportunityID
        INNER JOIN
        Course_Details as CorD
        ON ConD.SfCourseCID = CorD.SfCourseCID
-- Filter for entries that have grades, removing blank
"""
#creating pandas data frame from the joined 3 tables
data = pd.read_sql_query(myquery4, con=engine)
df = pd.DataFrame(data)


# adding in course pass fail though replace map
# Passing grades are A, A-, B, B+, B-, C, C-, Passed, S, SP
# Failing grades are FA, W, F, D, D+, D-, Failed, WP, FW, IF, WA, I, U

replace_map = {"A":1,
               "A-":1,
               "B":1, 
               "B+":1,
               "B-":1,
               "C":1,
               'C+':1,
               "C-":1,
               "Passed":1,
               "S":1,
               "SP":1,
               "FA":0,
               "W":0,
               "F":0,
               "D":0,
               "D+":0,
               "D-":0,
               "Failed":0,
               "WP":0,
               "FW":0,
               "IF":0,
               "WA":0,
               "I":0,
               "U":0,
               "":""}
df['PassFail'] = df['SfCourseCFinalGradeC'].replace(replace_map)
# Check unique values
# df['PassFail'].unique()
```

# CatBoost Model 
We are working with a highly categorical tabular data set leading me to use CatBoost. Other models may perform well on this data set but because of the aproximate 3 hour time limited for this project I have chosen to use CatBoot.CatBoost's ability to deal with categorical var will save a significant amount of time.


```python
#check data types catboost is sensitive to data types
model_df.dtypes
```

```python
#split data just for training and testing 
y = model_df['PassFail']
x = model_df.loc[:, model_df.columns != 'PassFail']

#Imputing missing values for both train and test
#needed for the format of catboost
x = x.fillna(-999)

#Split data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(x, y, train_size=0.6, random_state=42)
           
```

```python
#Identify categorical features indexs for model
categorical_features_indices = np.where(x.dtypes != np.float)[0]

#Create model 
#using mostly base model 
model = CatBoostClassifier(iterations=500,
                           learning_rate=.1,
                           depth=2,
                          logging_level = 'Silent')

# Fit model
model.fit(X_train, Y_train, cat_features=categorical_features_indices)
# Get predicted classes
preds_class = model.predict(X_test)
```

```python
# calculate precision
precision = precision_score(Y_test, preds_class, average='binary')
print('Precision: %.3f' % precision)
# calculate recall
recall = recall_score(Y_test, preds_class, average='binary')
print('Recall: %.3f' % recall)
#calculate accuracy
accuracy = accuracy_score(Y_test, preds_class)
print('Accuracy: %.3f' % accuracy)
#calculate f1
f1_metric = f1_score(Y_test, preds_class, average = "macro")
print('F1 Score: %.3f' % f1_metric)

#Confusion Matrix
print('\nConfusion Matrix')
confusion_matrix(Y_test,preds_class)
```

```python
#feature importance 
# code for feature importance plot from catboost  https://www.kaggle.com/mistrzuniu1/tutorial-eda-feature-selection-regression

feature_score = pd.DataFrame(list(zip(x.dtypes.index, model.get_feature_importance(Pool(x, label=y, cat_features=categorical_features_indices)))),
                columns=['Feature','Score'])

feature_score = feature_score.sort_values(by='Score', ascending=False, inplace=False, kind='quicksort', na_position='last')
plt.rcParams["figure.figsize"] = (12,7)
ax = feature_score.plot('Feature', 'Score', kind='bar', color='c')
ax.set_title("Catboost Feature Importance", fontsize = 14)
ax.set_xlabel('')
ax.set_xticklabels(feature_score.Feature, fontsize=16)
rects = ax.patches

labels = feature_score['Score'].round(2)

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 0.35, label, ha='center', va='bottom')


plt.show()
```
